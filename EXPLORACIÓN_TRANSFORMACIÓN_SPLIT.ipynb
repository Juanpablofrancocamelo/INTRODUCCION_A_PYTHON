```python
columnas=['Overall Qual', 'Overall Cond', 'Gr Liv Area',
           'Central Air', 'Total Bsmt SF', 'SalePrice']
 
df=pd.read_csv('http://jse.amstat.org/v19n3/decock/AmesHousing.txt',sep='\t' ,usecols=columnas)
df
```

```python
df["Central Air"].unique()
```

```python
df['Central Air']=np.where(df['Central Air']=='Y',1,0)
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna(axis=0)
df
```

```python
correlacion=df.corr(method="pearson")
correlacion
```

```python
X=df[["Overall Qual"]]
y=df["SalePrice"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
stdsc=StandardScaler()
X_train_std=stdsc.fit_transform(X_train)
X_test_std=stdsc.transform(X_test)
```

```python
X_train_std
```

```python
from sklearn import linear_model
reg=linear_model.LinearRegression()
reg=reg.fit(X_train_std,y_train)
reg
```

```python
print(f"Pendiente:{reg.coef_[0]}")
print(f"Intercepto: {reg.intercept_}")
```

```python
def linear_model(X,y,model):
    plt.scatter(X,y,c="steelblue")
    plt.plot(X,model.predict(X),color="black")
    return
```

```python
linear_model(X_train_std,y_train,reg)
plt.show()
```
